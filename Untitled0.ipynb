```python
# 1. Install dan import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.preprocessing import StandardScaler

# 2. Load data
file_path = '/content/heart_failure_clinical_records_dataset2.csv'  # pastikan file diupload ke Colab
data = pd.read_csv(file_path)

# 3. Pisahkan fitur dan target (misal 'DEATH_EVENT' sebagai target)
X = data.drop(columns=['DEATH_EVENT'])
y = data['DEATH_EVENT']

# 4. Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Terapkan LLE (misal jadi 2 dimensi)
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42)
X_lle = lle.fit_transform(X_scaled)

# 6. Visualisasikan hasilnya
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title('Local Linear Embedding (LLE)')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(scatter, label='DEATH_EVENT')
plt.grid(True)
plt.show()
```
